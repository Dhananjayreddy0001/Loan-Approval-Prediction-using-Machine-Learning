```python
import pandas as pd

```

```python
df=pd.read_csv("train.csv")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])

# Fill null values in 'Credit_History' column with the most common value
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])

# Fill null values in 'Married' and 'Dependents' columns with appropriate values
df['Married'] = df['Married'].fillna('No')  # Assuming default value is 'No' for missing values
df['Dependents'] = df['Dependents'].fillna('0') 

df['Loan_Status'] = df['Loan_Status'].fillna('N')  
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
df.isnull().sum()
```

```python
dataset=df.dropna()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset
```

```python


# Mapping dictionary for replacement
gender_mapping = {'Male': 1, 'Female': 0}

# Replace values in the 'Gender' column using the mapping
dataset['Gender'] = dataset['Gender'].map(gender_mapping)
```

```python
gender_mapping = {'Yes': 1, 'No': 0}

# Replace values in the 'Gender' column using the mapping
dataset['Married'] = dataset['Married'].map(gender_mapping)
```

```python
gender_mapping = {'Graduate': 1, 'Not Graduate': 0}

# Replace values in the 'Gender' column using the mapping
dataset['Education'] = dataset['Education'].map(gender_mapping)
```

```python
Self_Employed_mapping = {'Yes': 1, 'No': 0}

# Replace values in the 'Gender' column using the mapping
dataset['Self_Employed'] = dataset['Self_Employed'].map(Self_Employed_mapping)
```

```python
Loan_Status_mapping = {'Y': 1, 'N': 0}
dataset['Loan_Status'] = dataset['Loan_Status'].map(Loan_Status_mapping)
```

```python
x=dataset.drop('Loan_Status',axis=1)
```

```python
x.head()
```

```python
y=dataset['Loan_Status']
```

```python
y.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model_params = {

    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [50,60,70]
        }
    },
    
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini','entropy'],
        }
            
        },
     'XGBOOST': {
        'model':GradientBoostingClassifier(),
        'params': {
            'n_estimators': [100,200,130]
        }
            
        },
   
          
}

scores = []

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train,y_train)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

```python
model=GradientBoostingClassifier(n_estimators=100)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_pred,y_test)
```

```python
import seaborn as sb
sb.heatmap(cm,annot=True)
```

```python
print(classification_report(y_pred,y_test))
```
